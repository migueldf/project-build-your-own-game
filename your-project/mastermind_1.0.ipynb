```python
import random

colors = ["RED", "GREEN", "YELLOW", "BLUE", "BLACK", "PINK"]

colors_code = {1:"RED", 2:"GREEN", 3:"YELLOW", 4:"BLUE", 5:"BLACK", 6:"PINK"}

CPU_choice = random.sample(colors,4)
    
chances = 8

secret_code = ['???', '???', '???', '???']

player_attempts = [['---', '---', '---', '---'] for x in range(chances)]

hints = [['-', '-', '-', '-'] for x in range(chances)]


turn = 0


# funcion para imprimir tablero de juego
def board(CPU_choice, guess_codes, guess_flags):


    print("          Richo's & Mike's MASTERMIND")
    print("*"*50)

    print("        ||", end="")
    for x in CPU_choice:
        print("    " + x[:3], end="")
    print()	

    for i in range(len(guess_codes)):
        print("*"*50)

        print(hints[i][0], hints[i][1], hints[i][2], hints[i][3], end=" ||")
        for x in player_attempts[i]:
            print("    " + x[:3], end="")

        print()
    print("*"*50)



# Loop principal del juego
while turn < chances:

    print("1: Red, 2: Green, 3: Yellow, 4: Blue, 5: Black, 6: Pink")
    print("-"*50)
    print()
    board(secret_code, player_attempts, hints)

    # Ingresar valores de usuario 
    try:
        player_input = list(map(int, input("Enter your choice. Number between 1 and 6 separated by blank. No duplicates:").split()))
        
    # Check de ints 
    except ValueError:
        print("Error: Please enter a list of numbers.")
        continue   
 
    # Check de 4 numeros
    if len(player_input) != 4:
        print("Error: Please enter 4 numbers.")
        continue
    
    # Check de duplicados
    if len(player_input) != len(set(player_input)):
        print("Error: Please avoid duplicates and re-insert your choice.")
        continue
         
    # Check de colores validos
    error = 0
    for x in player_input:
            if x > 6 or x < 1:
                error = 1
     
    if error == 1:           
        print("Error: Please enter a number from the provided list.")
        continue

    # Se pasa todos los checks, se guarda variable en intentos.
    for i in range(4):
        player_attempts[turn][i] = colors_code[player_input[i]]

    # Lista temporal para destruir en proceso de verificación
    temp_CPU_choice = [x for x in CPU_choice]

    pos = 0

    # Asginar pins de pista
    
    #Copia de lista cpu y jugador. Si es correcto, se elimina de las listas.
    for x in player_input:
        if colors_code[x] in temp_CPU_choice:
            if player_input.index(x) == CPU_choice.index(colors_code[x]):
                hints[turn][pos] = 'R'
            else:
                hints[turn][pos] = 'W'
            pos += 1
            temp_CPU_choice.remove(colors_code[x])

    random.shuffle(hints[turn])


    # Condición para ganar
    if player_attempts[turn] == CPU_choice:
        board(CPU_choice, player_attempts, hints)
        print("You won!")
        break
        if input("Quieres continuar?") == "yes":
            continue

    turn += 1

# Condicion de perder
if turn == chances:
    board(CPU_choice, player_attempts, hints)
    print("You ran out of turns and lose!")
```
